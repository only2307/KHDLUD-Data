## Hàm mapping cho thuộc tính phân loại

```python
def gender_map(s):
  if s == 'M':
    s = -1
  else:
    s = 1
  return s
```

```python
def education_lev_map(s):
  if s == "Unknown":
    s = -1
  elif s == "Uneducated":
    s = 0
  elif s == "High School":
    s = 1
  elif s == "College":
    s = 2
  elif s == "Graduate":
    s = 3
  elif s == "Post-Graduate":
    s = 4
  else:
    s = 5
  return s
```

```python
def marital_status_map(s):
  if s == "Unknown":
    s = -1
  elif s == "Single":
    s = 0
  elif s == "Married":
    s = 1
  else:
    s = 2
  return s
```

```python
def cate_income_map(s):
  if s == "Unknown":
    s = -1
  elif s == 'Less than $40K':
    s = 0
  elif s == '$40K - $60K':
    s = 1
  elif s == '$60K - $80K':
    s = 2
  elif s == '$80K - $120K':
    s = 3
  else:
    s = 4
  return s      
```

```python
def card_map(s):
  if s == "Blue":
    s = 1
  elif s == "Silver":
    s = 2
  elif s == "Gold":
    s = 3
  else:
    s = 4
  return s 
```

```python
def target_map(s):
  if s == "Existing Customer":
    s = 1
  else:
    s = 0
  return s
```

## Sử dụng hàm mapping vào thuộc tính phân loại

```python
data['Gender'] = data['Gender'].apply(gender_map)
data["Education_Level"] = data["Education_Level"].apply(education_lev_map)
data["Marital_Status"] = data["Marital_Status"].apply(marital_status_map)
data["Income_Category"] = data["Income_Category"].apply(cate_income_map)
data["Card_Category"] = data["Card_Category"].apply(card_map)
data["Attrition_Flag"] = data["Attrition_Flag"].apply(target_map)
```
